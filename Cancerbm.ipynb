# **Step 4: Modelling**



```python
# Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

# **Regresssion**

```python

# Setting up feature matrix (X) and target vector (y) for regression
X_regression = clean_cp_no_outliers_numeric.drop(columns=['Cancer deaths attributable to alcohol_Proportion (%) of cancer deaths caused by alcohol drinking in men ages 15 years or older, 2016'])
y_regression = clean_cp_no_outliers_numeric['Cancer deaths attributable to alcohol_Proportion (%) of cancer deaths caused by alcohol drinking in men ages 15 years or older, 2016']


# Split the data for regression and classification
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.2, random_state=42)


# 1. Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_reg, y_train_reg)
y_pred_lin = lin_reg.predict(X_test_reg)
mse_lin = mean_squared_error(y_test_reg, y_pred_lin)
r2_lin = r2_score(y_test_reg, y_pred_lin)


# 2. Random Forest Regression
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train_reg, y_train_reg)
y_pred_rf = rf_reg.predict(X_test_reg)
mse_rf = mean_squared_error(y_test_reg, y_pred_rf)
r2_rf = r2_score(y_test_reg, y_pred_rf)


```

```python
# Get feature importances
feature_importances = rf_reg.feature_importances_

# Visualize feature importances
plt.figure(figsize=(10, 6))
plt.barh(X_train_reg.columns, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance Scores')
plt.show()
```

# **Classification**

```python

# Setting up feature matrix (X) and target vector (y) for classification
X_classification = clean_cp_no_outliers_numeric.drop(columns=['Cervical cancer incidence rates_Age-standardized rate (world) per 100,000, 2018'])
y_classification = clean_cp_no_outliers_numeric[ 'Cervical cancer incidence rates_Age-standardized rate (world) per 100,000, 2018']


# Binarizing classification target (for classification model)
clean_cp_no_outliers_numeric['Cervical_Cancer_High_Low'] = (clean_cp_no_outliers_numeric[ 'Cervical cancer incidence rates_Age-standardized rate (world) per 100,000, 2018'] >clean_cp_no_outliers_numeric[ 'Cervical cancer incidence rates_Age-standardized rate (world) per 100,000, 2018'].median()).astype(int)
y_classification = clean_cp_no_outliers_numeric['Cervical_Cancer_High_Low']



# Split the data for regression and classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_regression, y_classification, test_size=0.2, random_state=42)

# Initialize the LabelEncoder
le = LabelEncoder()
le.fit(y_test_class)

# 1. Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_class, y_train_class)
y_pred_log = log_reg.predict(X_test_class)
accuracy_log = accuracy_score(y_test_class, y_pred_log)

# 2. Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train_class, y_train_class)
y_pred_rf_class = rf_clf.predict(X_test_class)
accuracy_rf_class = accuracy_score(y_test_class, y_pred_rf_class)



```

# **Step 5: Evaluation**

```python
# Evaluation for Regression
models = ['Linear Regression', 'Random Forest']
predictions = [y_pred_lin, y_pred_rf]
results = {}

for model, pred in zip(models, predictions):
    mse = mean_squared_error(y_test_reg, pred)
    r2 = r2_score(y_test_reg, pred)
    results[model] = {'MSE': mse, 'R2 Score': r2}

results_df = pd.DataFrame(results).T
print(results_df)
```

Random Forest outperforms Linear Regression based on both MSE and RÂ².

 Random Forest is better at reducing prediction error (lower MSE) and better at explaining variance (higher RÂ²).

The relatively low RÂ² scores (0.293 and 0.389) for both models indicate that neither model is explaining the majority of the variance in the target variable.

```python
# Evaluation for Classification
classification_models = ['Logistic Regression', 'Random Forest']
classification_predictions = [y_pred_log, y_pred_rf_class, ]
classification_results = {}

for model, pred in zip(classification_models, classification_predictions):
    report = classification_report(y_test_class, pred, output_dict=True)
    classification_results[model] = {
        'Precision': report['weighted avg']['precision'],
        'Recall': report['weighted avg']['recall'],
        'F1 Score': report['weighted avg']['f1-score']
    }

classification_results_df = pd.DataFrame(classification_results).T
print(classification_results_df)

```

Random Forest fit the model perfectly at 100% followed by Logistic Regression at 93%

```python
#Classification Predictions vs Actual

# Confusion Matrix for each classifier
models = ['Logistic Regression', 'Random Forest']
predictions = [y_pred_log, y_pred_rf_class]

# Create subplots for each confusion matrix
fig, axes = plt.subplots(2, 1, figsize=(12, 12))
axes = axes.flatten()

for ax, model, pred in zip(axes, models, predictions):
    cm = confusion_matrix(y_test_class, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=le.classes_, yticklabels=le.classes_)
    ax.set_title(f'Confusion Matrix - {model}')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.tight_layout()
plt.show()
```
