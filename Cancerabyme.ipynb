```python
# Outliers
outliers = newnumericc.apply(lambda x: (x - x.mean()).abs() > 3 * x.std()).sum()

print("\nOutliers:")
print(outliers)
```

It is only Lung cancer incidence rates, male\nAge-standardized rate (world) per 100,000, all ages, 2018 column that has no outlier                                             

```python
# Define the z-score threshold for outlier detection
threshold = 3

# Calculate z-scores for each numerical column
z_scores = newnumericc.apply(lambda x: (x - x.mean()) / x.std())

# Find rows where any z-score exceeds the threshold
outlier_rows = z_scores[(z_scores.abs() > threshold).any(axis=1)]
outlier_rows
```

```python
# Removing Outliers
newnumericc_filtered = newnumericc.drop(outlier_rows.index)

# Display the filtered DataFrame
newnumericc_filtered
```

```python
 #Missing Values Check
missing_values =newnumericc.isnull().sum()
print(missing_values)
```
